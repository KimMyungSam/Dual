### 종목코드를 0으로 채우고 6자리로 만듬

```python
code = df_krx["Symbol"].apply(lambda x:format(x, "06d"))
```

```python
code.head()
```

### 전체 종목의 가격 data 가져오고 concat

```python
import sys
```

```python
total = pd.Series()
```

```python
for name in code:
    result = fdr.DataReader(name)
    result = result["Close"]
    result.name = name
    total = pd.concat([total,result], axis=1)
    sys.stdout.write(str(name)+", ")
```
